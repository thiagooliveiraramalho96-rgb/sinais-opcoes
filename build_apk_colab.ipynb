```python
# Inicializar Buildozer se não existir o .spec
import os
if not os.path.exists('buildozer.spec'):
    !buildozer init
    print('✅ buildozer.spec criado')
else:
    print('✅ buildozer.spec já existe')
```

```python
# Configurar buildozer.spec para o ambiente Colab
# Ajustar paths e requisitos
!sed -i 's/requirements = python3,kivy==2.3.1/requirements = python3,kivy==2.3.1,kivymd==1.2.0,pandas==2.2.2,numpy==2.2.6,yfinance==1.4.1,pandas_ta==0.4.71b0,matplotlib==3.10.9,plyer==2.1.0,schedule==1.2.2,scipy==1.15.2,requests==2.32.3/' buildozer.spec
!sed -i 's/android.api = .*/android.api = 33/' buildozer.spec
!sed -i 's/android.minapi = .*/android.minapi = 21/' buildozer.spec
!sed -i 's/#android.ndk = .*/android.ndk = 27b/' buildozer.spec
!sed -i 's/osx.python_version = .*/osx.python_version = 3.12/' buildozer.spec
!sed -i 's/package.name = .*/package.name = sinaisopcoes/' buildozer.spec
!sed -i 's|package.domain = .*|package.domain = org.trade.b3|' buildozer.spec
!sed -i 's/title = .*/title = Sinais Opções B3/' buildozer.spec
print('✅ buildozer.spec configurado')
```

```python
# ⚠️ BUILD DO APK - Pode levar 30-60 minutos!
print('🚀 Iniciando build do APK...')
print('   Isso pode levar até 1 hora no primeiro build')
print('   (Baixando NDK, SDK, dependências...)\n')

!buildozer android debug 2>&1 | tail -50
```

```python
# Verificar se o APK foi gerado
!ls -la bin/*.apk 2>/dev/null || echo '❌ APK não encontrado'
```

```python
# Baixar o APK
from google.colab import files
import glob

apk_files = glob.glob('bin/*.apk')
if apk_files:
    for apk in apk_files:
        print(f'📦 APK gerado: {apk}')
        files.download(apk)
else:
    print('❌ Nenhum APK encontrado. Verifique os erros acima.')
```
